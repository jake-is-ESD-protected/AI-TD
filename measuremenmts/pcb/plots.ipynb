```python
"""
NOTE
----

The measurements were done using a 1/12 octave filterbank as replacement for an FFT because of the
limits of the measurement hardware (NTi-Audio XL2). This means that for the assumption of electrical
noise in the form of white noise, the exponentially growing bandwidths of the filterbank create an
inherent rise in total power per band because more frequencies are populated there. To fix this,
we employ a correction factor `corr_12` based on bandwidth which scales the amplitudes accordingly.
In this representation, white noise produces a roughly "flat" amplitude response similar to the
output of an FFT.
"""

ref_0dbu = np.max(M2_220HZ_W_12OCT)
corr_12 = 10 * np.log10(np.array(freqs_12oct) * (np.sqrt(2)**(1/24) - 1))

plt.figure(figsize=(10, 6))
plt.plot(freqs_12oct, M1_OFF_R_12OCT - corr_12 -ref_0dbu, label="Ruhebetrieb rechter Kanal")
plt.plot(freqs_12oct, M1_OFF_W_12OCT - corr_12 - ref_0dbu, label="Ruhebetrieb linker Kanal")
plt.plot(freqs_12oct, M2_220HZ_R_12OCT - corr_12 - ref_0dbu, label="Aktives Gert rechter Kanal")
plt.plot(freqs_12oct, M2_220HZ_W_12OCT - corr_12 - ref_0dbu, label="Aktives Gert linker Kanal")
plt.semilogx()
plt.title("Spektrum am Gerteausgang mit 0 dBu 220 Hz Sinussignal")
plt.xlabel("Frequenz in Hz")
plt.ylabel("Pegel in dBu")
plt.grid(which='both')
plt.legend()
```

```python
"""
NOTE
----

The sweep measurement done here (NTi-Audio MR-Pro) was of exponential nature, which means that more time
and therefore energy was spent in lower frequencies. This directly negates the inherent amplitude rise 
over frequency in the 1/12 octave filterbank measurement, meaning that a correction is not needed in order
to show a "flat" response.
"""

plt.figure(figsize=(10, 6))

sum1 = np.power(10, np.asarray(M3_SWEEP_R_6_OCT_N1)/20) + np.power(10, np.asarray(M3_SWEEP_R_6_OCT_N2)/20)
sum2 = np.power(10, np.asarray(M3_SWEEP_W_6_OCT_N1)/20) + np.power(10, np.asarray(M3_SWEEP_W_6_OCT_N2)/20)

M3_SWEEP_R_6_OCT = 20*np.log10(sum1/2)
M3_SWEEP_W_6_OCT = 20*np.log10(sum2/2)

plt.plot(freqs_6oct, M3_SWEEP_R_6_OCT-ref_0dbu, label="Amplitudengang mit Sinus-Sweep rechter Kanal")
plt.plot(freqs_6oct, M3_SWEEP_W_6_OCT-ref_0dbu, label="Amplitudengang mit Sinus-Sweep linker Kanal")
plt.semilogx()
plt.title("Spektrum am Gerteausgang mit Sinus-Sweep")
plt.xlabel("Frequenz in Hz")
plt.ylabel("Pegel in dBu")
plt.grid(which='both')
plt.legend()
```
